```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'C:\Users\umair\OneDrive\Desktop\My Files\ExcelR\Assignments\Association Rules\book.csv')
df
```

```python
score = apriori(df,min_support=0.1,use_colnames=True)
score
```

```python
model = association_rules(score,metric='lift')
model
```

```python
model[model.lift>1]
```

```python
sns.scatterplot(data=model, x="support", y="confidence")
```

```python
model[['support','confidence','lift']].hist()
```

```python
sns.scatterplot(data=model, x="support", y="confidence",hue='antecedents')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
